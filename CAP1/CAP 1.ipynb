## Create Thumbnails 

Using PIL to create thumbnails is very simple. The **thumbnail()** method takes a tuple specifying the new size and converts the image to a thumbnail image with size that fits within the tuple. To create a thumbnail with longest side 128 pixels, use the method like this:

```python
pil_img.thumbnail((128,128))
```

```python
imshow(pil_img)
```

## Copy and Paste Regions

Cropping a region from an images is done using the **crop()** method:

```python
box = (100,100,400,400)
region = pil_img.crop(box)
```

The region is defined a 4-Tuple, where coordinates are (left,upper,right,lower). PIL uses a coordinate system with (0,0) in the upper left corner. The extracted region can, for example, be rotated and then put back using the **paste()** method lihe this:

```python
region = region.transpose(Image.ROTATE_180)
pil_img.paste(region,box)
```

## Resize and Rotate 

To resize an image, call **resize()** with a tu8ple giving the new size:

```python
out = pil_img.resize((128,128))
```

To rotate an image, use counterclockwise angles and **rotate()** like this:

```python
out = pil_img.rotate(45)
imshow(out)
```

## Image Contours and Histograms 

```python
im = np.array(Image.open('nana.png').convert('L'))

# create a new figure
figure()
#don't use colors
gray()
#show contours with origin upper left corner
contour(im,origin='image')
axis('equal')
axis('off')

figure()
hist(im.flatten(),128)
show()
```

An image histogram is a plot showing the distribution of pixel values. A number of bins is specified for the span of values and each bin gets a count of how many pixels have values in the bin's range. The visualization of the (graylevel) image histogram is done using the **hist()** function.
